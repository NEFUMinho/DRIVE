(a) Mostre que o sistema  descrito pelo seguinte sistema de equaes, onde  = 2k  m2:
$ \\
(  k)x_1  kx_2 = C \\
x_i  kx_i-1  kx_i+1 = 0 \\
(  k)x_N  kx_N1 = 0 \\
$

Sendo:  
$ \\
_{i}(t)' = iwx_{i}e^{iwt} $

$_{i}(t)'' = -w^{2}x_{i}e^{iwt} $

Substitumos os valores das derivadas nas equaes:
$ \\
m{d^{2}_{i} \over dt^{2}} = k(_{i+1} - _{i}) + k(_{i-1} - _{i}) + F_{i} (1)\\
m{d^{2}_{i} \over dt^{2}} = k(_{2} - _{1}) + F_{i} (2)\\
m{d^{2}_{N} \over dt^{2}} = k(_{N-1} - _{N}) + F_{N} (3) $

Substituindo os valores em (1) e sabendo que nenhuma fora est a ser aplicada s massas intermdias, temos:

$ \\
-mw^{2}x_{i}e^{iwt} = kx_{i+1}e^{iwt}-kx_{i}e^{iwt} + kx_{i-1}e^{iwt} -kx_{i}e^{iwt} <=> $

$ <=> -mw^{2}x_{i} = kx_{i+1} -2kx_{i} + kx_{i-1} <=> $

$<=> (2k-mw^{2})x_{i} -kx_{i+1} - kx_{i-1} = 0 $

$ <=> x_{i} - kx_{i+1} - kx_{i-1} = 0 $

Substituindo os valores em (2) e sabendo que a fora aplicada  massa 1  do tipo $ F_1=ce^{iwt}$, temos:

$ \\
-mw^{2}x_{1}e^{iwt} = k(x_{2}e^{iwt}-x_{1}e^{iwt}) + ce^{iwt} <=> $

$ <=> -mw^{2}x_{1} = k(x_{2}-x_{1}) + c <=> $

$ <=> -mw^{2}x_{1} - kx_{2} +kx_{1} = c <=> $

$<=> (k - mw^{2})x_{1} -kx_{2} = c $

Substituindo os valores em (3) e sabendo que nenhuma fora est a ser aplicada  massa N, temos:

$ \\
-mw^{2}x_{N}e^{iwt} = k(x_{N-1}e^{iwt}-x_{N}e^{iwt}) <=> $

$ <=> -mw^{2}x_{N} = k(x_{N-1}-x_{N}) <=> $

$ <=> -mw^{2}x_{N} - kx_{N-1} +kx_{N} = 0 <=> $

$<=> (k - mw^{2})x_{N} -kx_{N-1} = 0 $


(b) Resolva este problema para N = 26, C = 1, m = 1, k = 6 e  = 2.

```python
import numpy as np
def eq_to_mat(N,C,m,k,w):                                   #definimos a equao que recebe os nossos dados e cria uma matriz
    a = 2*k-m*(w**2)                                        #definimos a varivel alpha
    lista = []                                              #definimos 'lista', que ser a nossa matriz final
    listt = []                                              #definimos 'listt', que ajudar a criar cada linha da matriz    
    lista.append([a-k,-k])                                  #adicionamos a lista [a-k,-k] a 'lista', que so os primeiros coeficientes da primeira equao
    for b in range(N-2):                                    #os restantes coeficientes (ou seja, N-2) so zeros
        lista[0].append(0)                                  #adicionamos ao primeiro elemento de 'lista' (lista[0]) os zeros em falta
    lista[0].append(C)                                      #por fim, adicionamos C, a constante  qual a equao est igualada
    for i in range(1,N-1):                                  #na prxima poro de cdigo, adicionamos a 'lista' as linhas compreendidas entre a primeira e a ltima linhas
        for b in range(0,i-1):                              #adicionamos os 0 que esto antes dos coeficientes em questo (depende da linha a ser trabalhada)
            listt.append(0)
        listt.append(-k)                                    #adicionamos o coeficiente de x_i-1
        listt.append(a)                                     #adicionamos o coeficiente de x_i
        listt.append(-k)                                    #adicionamos o coeficiente de x_i+1
        for b in range(i+1,N):                              #adicionamos os 0 que esto depois dos coeficientes em questo (de i+1 at N, porque a equao est igualada a 0, ento adicionamos a ltima constante)
            listt.append(0)
        lista.append(listt)                                 #depois de criada a linha da matriz, esta  adicionada  lista que representa a matriz
        listt = []                                          #voltamos a igualar listt a [], para que, ao percorrer novamente o ciclo, possamos criar a linha seguinte
    for b in range(N-2):                                    #na prxima poro de cdigo vamos fazer a ltima linha da matriz
        listt.append(0)                                     #adicionamos os zeros antes dos coeficientes em questo
    listt.append(-k)                                        #adicionamos o coeficiente de x_N-1
    listt.append(a-k)                                       #adicionamos o coeficiente de x_N
    listt.append(0)                                         #adiconamos o 0  qual a equao est igualada
    lista.append(listt)                                     #adicionamos a 'lista' a ltima coluna da matriz
    return lista                                            #a funo devolve a matriz

#Teste da funo
print(np.array(eq_to_mat(26,1,1,6,2)))
```

```python
def metodo(A):                                                 #definimos a funo que vai aplicar o mtodo de gauss
    solucoes=[]                                                #criamos uma lista na qual vo ser armazenadas as solues
    for i in range(len(A)):                                    #a varivel i vai dar a linha em que estamos a trabalhar
        for m in range(i+1, len(A)):                           #a varivel m vai dar a linha abaixo de i       
            mult = A[m][i]/A[i][i]                             #calculamos o fator multiplicativo do mtodo de gauss, dado por a_mn/a_nn, sendo a_nn o elemento piv de cada linha e a_mn o elemento imediatamente abaixo dele                                        
            for n in range (i, len(A)+1):                      #a varivel n vai percorrer todos os elementos de cada linha a partir de i
                A[m][n] = A[m][n]- mult*A[i][n]                #aplicamos a transformao L_m = L_m - (a_mn/a_nn)*L_n a cada elemento da respetiva linha (seguindo a terminologia dos comentrios e no do cdigo), por forma a obter uma matriz triangular
    
    solucoes = np.zeros(len(A))                                #vai armazenar na lista vazia o nmero de solues do sistema (igual ao nmero de linhas da matriz) enquanto zeros
    linha = len(A)-1                                           #a varivel linha guarda a linha em que estamos a trabalhar
    b = len(A)                                                 #a varivel b guarda o nmero da ltima coluna, que contm os nmeros aos quais as expresses esto igualadas 
    while linha>=0:                                            #como vamos decrementar 'linha', impomos esta condio. comeamos o ciclo na ltima linha
        x = A[linha][b]                                        #atribuimos  varivel x o valor contido na ltima coluna, na respetiva linha a ser trabalhada
        coluna = len(A)-1                                      #a varivel coluna guarda a coluna em que estamos a trabalhar (exclumos a ltima porque no  trabalhada)
        while coluna>linha:                                    #entramos no ciclo se estivermos a trabalhar numa posio da matriz acima da diagonal criada na triangularizao. no primeiro caso a ser trabalhado (ltima linha e penltima coluna), no entramos no ciclo porque linha>coluna, armazenando logo o valor de x_m na ltima posio da lista solucoes
            x = x - A[linha][coluna]*solucoes[coluna]          #recorrendo  frmula do mtodo de Gauss
            coluna = coluna -1                                 #decrementamos o valor da coluna em 1
        x = x/A[linha][linha]                                  
        linha = linha -1                                       #decrementamos o valor da linha em 1 para trabalhar a equao da linha anterior
        solucoes[coluna]=x                                     #igualamos o valor da soluo para a coluna em que estamos a trabalhar a x.

    return solucoes
```

```python
from numpy import copy
#Alteramos a primeira funo, de forma a que a ltima coluna no exista, para que sirva de argumento  funo banded

def eq_to_mat2(N,C,m,k,w):                                  #definimos a equao que recebe os nossos dados e cria uma matriz
    a = 2*k-m*(w**2)                                        #definimos a varivel alpha
    lista = []                                              #definimos 'lista', que ser a nossa matriz final
    listt = []                                              #definimos 'listt', que ajudar a criar cada linha da matriz    
    lista.append([a-k,-k])                                  #adicionamos a lista [a-k,-k] a 'lista', que so os primeiros coeficientes da primeira equao
    for b in range(N-2):                                    #os restantes coeficientes (ou seja, N-2) so zeros
        lista[0].append(0)                                  #adicionamos ao primeiro elemento de 'lista' (lista[0]) os zeros em falta

    for i in range(1,N-1):                                  #na prxima poro de cdigo, adicionamos a 'lista' as linhas compreendidas entre a primeira e a ltima linhas
        for b in range(0,i-1):                              #adicionamos os 0 que esto antes dos coeficientes em questo (depende da linha a ser trabalhada)
            listt.append(0)
        listt.append(-k)                                    #adicionamos o coeficiente de x_i-1
        listt.append(a)                                     #adicionamos o coeficiente de x_i
        listt.append(-k)                                    #adicionamos o coeficiente de x_i+1
        for b in range(i+1,N-1):                            #adicionamos os 0 que esto depois dos coeficientes em questo (de i+1 at N-1, porque no queremos a ltima coluna)
            listt.append(0)
        lista.append(listt)                                 #depois de criada a linha da matriz, esta  adicionada  lista que representa a matriz
        listt = []                                          #voltamos a igualar listt a [], para que, ao percorrer novamente o ciclo, possamos criar a linha seguinte
    for b in range(N-2):                                    #na prxima poro de cdigo vamos fazer a ltima linha da matriz
        listt.append(0)                                     #adicionamos os zeros antes dos coeficientes em questo
    listt.append(-k)                                        #adicionamos o coeficiente de x_N-1
    listt.append(a-k)                                       #adicionamos o coeficiente de x_N

    lista.append(listt)                                     #adicionamos a 'lista' a ltima linha da matriz
    
    return lista                                            #a funo devolve a matriz
 

def banded(Aa,va,up,down):

    # Copy the inputs and determine the size of the system
    A = copy(Aa)
    v = copy(va)
    N = len(v)

    # Gaussian elimination
    for m in range(N):

        # Normalization factor
        div = A[up,m]

        # Update the vector first
        v[m] /= div
        for k in range(1,down+1):
            if m+k<N:
                v[m+k] -= A[up+k,m]*v[m]

        # Now normalize the pivot row of A and subtract from lower ones
        for i in range(up):
            j = m + up - i
            if j<N:
                A[i,j] /= div
                for k in range(1,down+1):
                    A[i+k,j] -= A[up+k,m]*A[i,j]

    # Backsubstitution
    for m in range(N-2,-1,-1):
        for i in range(up):
            j = m + up - i
            if j<N:
                v[m] -= A[i,j]*v[j]

    return v

#A funo fun_f ser a funo final  qual ser dada apenas a funo obtida em eq_to_mat2. Esta ter como tarefa criar e atribur as variveis  banded.

def fun_f(A):
    #denominamos as variveis respectivas a este problema.
    k = 6
    m = 1
    w = 2
    a = 2*k-m*(w**2)
    N = len(A)                                              #N  igual ao tamanho de A
    
#Cramos a lista que ser dada na funo banded como o argumento 'va'. Esta lista  composta N elementos em que o primeiro  igual a C e os restantes a zero (neste problema, C=1) e  equivalente
#s igualdades das equaes iniciais.

    v = np.zeros(N,float)                                   #cramos lista com N elementos iguais a 0.
    v[0] = 1                                                #igualamos o primeiro elemento da lista a C que para este problema  1.

#a funo banded recebe uma matriz composta pelas 3 diagonais com elementos diferentes de 0 que compem a matriz obtida na funo eq_to_mat2 excluindo os 0 desnecessrios aos clculos.
#para este caso:

    A = np.empty([3,N],float)                               #cria-se 3 listas com N elementos  
    A[0,:] = -k                                             #os primeiros elementos de cada lista em A  igualado a -k
    A[1,:] = a                                              #os segundos elementos de cada lista em A  igualado a a
    A[2,:] = -k                                             #os terceiros elementos de cada lista em A  igualado a -k
    A[1,0] = a -k                                           #os elementos das extremidades das matrizes sabemos pelas equaes que o primeiro elemento da segunda linha  igual a a-k
    A[1,N-1] = a -k                                         #o ltimo elemento da segunda linha  igual a a-k

    resultado = banded(A,v,1,1)                             #o resultado final ser dado pela funo banded.
    
    return resultado

#Teste da funo fun_f:
fun_f(eq_to_mat2(1000,1,1,6,2))

#os resultados so apresentados imediatamente a seguir ao pedido de execuo da clula provando a melhor eficcia comparativamente ao mtodo anterior.
```
