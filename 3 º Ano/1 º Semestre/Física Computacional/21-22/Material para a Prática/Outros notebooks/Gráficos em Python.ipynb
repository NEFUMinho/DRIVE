```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(-x**2)*np.cos(20*x)

x = np.linspace(-2, 2, 50)

plt.plot(x,f(x))
plt.plot(x,f(x),'ro')
plt.ylabel('y = exp(-x**2)*cos(20*x)')
plt.xlabel('x')

plt.show()
```

```python
def f(x, y):
    return np.sin(x) * np.cos(y)

x = np.linspace(-3, 3, 30)
y = np.linspace(-3, 3, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 60, cmap='hot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```
