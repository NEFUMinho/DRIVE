```python
## exercicio 1

## np.empty([x,y], float) cria uma matriz  de tamanho x*y e com floats mas tudo a 0.0s


import numpy as np
import math
import matplotlib.pyplot as plt


side=100.0
points=500
spacing=side/points
a1=1
a2=1
k1=(2*np.pi)/5
k2=(2*np.pi)/5


xi=np.empty([points,points],float)

def psi1(x,y):
    return(a1*np.sin(k1*np.sqrt((x-40)**2+(y-50)**2)))

def psi2(x,y):
    return(a2*np.sin(k2*np.sqrt((x-60)**2+(y-50)**2)))

for i in range(points):
    y=spacing*i
    for j in range(points):
        x=spacing*j
        xi[i,j]=psi1(x,y)+psi2(x,y)

plot=plt.imshow(xi,origin="lower",extent=[0,side,0,side])
cb=plt.colorbar(plot)
cb.set_label("altura da agua [cm]")
plt.gray()
plt.show()
```

```python
#pergunta2

# x'=rx(1-x)
#x<1 e r não muito grandes   
#truques: plot normal faz pontos grandes demais, por isso usa-se plt(x,y,"g,")  ou seja pinta-se um ponto de verde e mete-se a virgula á frente do g para o ponto ser pequeno
import numpy as np
import math
import matplotlib.pyplot as plt


rmin=0
rmax=4
rsetp=0.001
steps=1000

r=np.arange(rmin,rmax,rsetp)
x=np.empty(len(r),float)
x[:]=0.5 #atribui a todos os elementos do array o valor 0.5
print(r)
print(x)

fig=plt.figure(figsize=(15,10))
#estas primeiras mil é so para estabilizar
for i in range(steps):
    x=r*x*(1-x)
    
 #estas já são para plotar   
for i in range(steps):
    x=r*x*(1-x)
    plt.plot(r,x,"g,")
    
    
plt.xlabel("R")
plt.ylabel("X")
    
```
