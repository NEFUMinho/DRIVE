```python
#Exercício 1
#imports
from math import sin
from numpy import arange
import pylab
#função
def f(x,t):
    return -x**3 + sin(t)
#condições
a=0.0
b=10.0
N=1000
h=(b-a)/N
x=0.0

tpoints=arange(a,b,h)
xpoints= []
for t in tpoints:
    xpoints.append(x)
    x+=h*f(x,t)
    
    
pylab.plot(tpoints,xpoints)
pylab.xlabel("t")
pylab.ylabel("x(t)")
pylab.show
```

```python
#imports
from numpy import arange
import pylab
#função
def f(x,t):
    return -x**3 + sin(t)


def d_2(N):
    a=0.0
    b=10.0
    x=0.0
    h=(b-a)/N
    tpoints=arange(a,b,h)
    xpoints= []
    for t in tpoints:
        xpoints.append(x)
        k1 = h*f(x,t)
        k2 = h*f(x+0.5*k1,t+0.5*h)
        x += k2
    return tpoints,xpoints
    
    

pylab.plot(d_2(10)[0],d_2(10)[1], label="10")
pylab.plot(d_2(20)[0],d_2(20)[1], label="20")
pylab.plot(d_2(50)[0],d_2(50)[1], label="50")
pylab.plot(d_2(100)[0],d_2(100)[1], label="100")
pylab.title("Segunda Ordem")
pylab.legend()
pylab.xlabel("t")
pylab.ylabel("x(t)")
pylab.show()


def d_4(N):
    a=0.0
    b=10.0
    x=0.0
    h=(b-a)/N
    tpoints=arange(a,b,h)
    xpoints= []
    for t in tpoints:
        xpoints.append(x)
        k1 = h*f(x,t)
        k2 = h*f(x+0.5*k1,t+0.5*h)    
        k3 = h*f(x+0.5*k2,t+0.5*h)
        k4 = h*f(x+k3,t+h)
        x += (k1+2*k2+2*k3+k4)*(1/6)
    return tpoints,xpoints




pylab.plot(d_4(10)[0],d_4(10)[1], label="10")
pylab.plot(d_4(20)[0],d_4(20)[1], label="20")
pylab.plot(d_4(50)[0],d_4(50)[1], label="50")
pylab.plot(d_4(100)[0],d_4(100)[1], label="100")
pylab.xlabel("t")
pylab.ylabel("x(t)")
pylab.title("Quarta Ordem")
pylab.legend()
pylab.show()
```
