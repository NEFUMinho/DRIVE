```python
from math import pi, sin, cos, sqrt
from pylab import plot, show, xlabel, ylabel
from numpy import arange, array

g=9.81
R=0.08
p=1.22
C=0.47
theta=pi/6
v_0=100
m=1
c=(pi*p*C*(R**2))/(2*m)
t_0=0
x_0=0
y_0=0
t_f=float(input("Qual Ã© o tempo final que pretende?:"))


def f(r, t):
    vx = r[1]
    vy = r[3]
    v = sqrt(vx ** 2 + vy ** 2)
    return array([vx,- c * vx * v, vy, -g - c * vy * v], float)





def graph(N):
    r = array([0, v_0 * cos(theta), 0, v_0 * sin(theta)], float)
    h=(t_f - t_0) / N
    tpoints = arange(t_0, t_f, h)
    xpoints = []
    ypoints = []
    for t in tpoints:
        xpoints.append(r[0])
        ypoints.append(r[2])
        k1 = h * f(r, t)
        k2 = h * f(r + 0.5 * k1, t + 0.5 * h)
        k3 = h * f(r + 0.5 * k2, t + 0.5 * h)
        k4 = h * f(r + k3, t + h)
        r += (k1 + 2 * k2 + 2 * k3 + k4) / 6
        #if r[2]<=0:
        #    break;
    return array(xpoints, float), array(ypoints, float)


for i in range (1,7):
    print(i)
    e=10**i
    plot(graph(e)[0], graph(e)[1])
    xlabel('x')
    ylabel('y')
    show()
    
    
#aqui verifiquei que H=100 Ã© suficiente, mas vou usar 1000 sÃ³ para estar sure. aqui usei t_f=10 e jÃ¡ passou o 0, logo Ã© diminuir.

```

```python
from math import pi, sin, cos, sqrt
from pylab import plot, show, xlabel, ylabel
from numpy import arange, array, linspace

g=9.81
R=0.08
p=1.22
m=1
theta=pi/6
v_0=100
t_0=0
x_0=0
y_0=0
t_f=7
N=1000
h=(t_f - t_0) / N
tpoints = arange(t_0, t_f, h)


def f(r, t, c):
    vx = r[1]
    vy = r[3]
    v = sqrt(vx ** 2 + vy ** 2)
    return array([vx,- c * vx * v, vy, -g - c * vy * v], float)





def graph(C):
    c=(pi*p*C*(R**2))/(2*m)
    r = array([0, v_0 * cos(theta), 0, v_0 * sin(theta)], float)
    xpoints = []
    ypoints = []
    for t in tpoints:
        xpoints.append(r[0])
        ypoints.append(r[2])
        k1 = h * f(r, t, c)
        k2 = h * f(r + 0.5 * k1, t + 0.5 * h, c)
        k3 = h * f(r + 0.5 * k2, t + 0.5 * h, c)
        k4 = h * f(r + k3, t + h, c)
        r += (k1 + 2 * k2 + 2 * k3 + k4) / 6
        #if r[2]<=0:
        #    break;
    
    return array(xpoints, float), array(ypoints, float)

cpoints = linspace(1,10,10)


for i in cpoints:
    print(i)
    plot(graph(i)[0], graph(i)[1])
    xlabel('x')
    ylabel('y')
    show()

cpoints = linspace(0.1,1,10)


for i in cpoints:
    print(i)
    plot(graph(i)[0], graph(i)[1])
    xlabel('x')
    ylabel('y')
    show()

#No C Ã© possÃ­vel ver precisamente o contrÃ¡rio do que vimos com m. A medida que C aumenta, diminui o y mÃ¡ximo, o "cume"
#a altura mÃ¡xima do canhÃ£o e diminuio tempo atÃ© voltar ao chÃ£o.
```
