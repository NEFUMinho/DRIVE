```python
%%time
#Ficha 8
#Exercício 4
#Miguel Godinho A89624
#André Mouquinho A90129
from numpy import zeros
from pylab import imshow, colorbar,set_cmap, show


V = 1.0
a = 0.1
N = 100
delta = 10 ** (-6)
omega = 0.9


max_diff = 1.0000005*delta

phi = zeros([N + 1, N + 1], float)
phi[20:81, 20] = V
phi[20:81, 80] = -V



while max_diff > delta:
    max_diff = 0.0
    for i in range(N + 1):
        for j in range(N + 1):
            if not i == 0 and not j == 0 and not i == N and not j == N and not (20 <= i <= 80 and j == 20 or j == 80):
                old_phi = phi[i,j]
                new_phi = (1 + omega) * (phi[i + 1, j] + phi[i - 1, j] + phi[i, j + 1] + phi[i, j - 1]) / 4 \
                          - omega * old_phi
                phi[i, j] = new_phi
    if max_diff >= abs(new_phi - old_phi):
        max_diff=max_diff
    else:
        max_diff=abs(new_phi - old_phi)
                    


imshow(phi)
colorbar()
set_cmap("Reds_r")
show()
```
