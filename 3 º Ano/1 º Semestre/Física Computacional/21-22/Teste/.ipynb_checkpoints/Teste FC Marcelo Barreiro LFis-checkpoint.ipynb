```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#definição de uma função que procede ao método de eliminação de Gauss
def Gauss(A,c):
    M=np.copy(A)
    b=np.copy(c)
    n = len(b)
    x=np.empty(n)
    #tr
    for k in range(0,n):
        M[k,:]=M[k,:]/M[k,k]
        b[k] /= M[k,k]
        for i in range(n-1,0+k,-1):
            a=M[i,k]
            b[i] -= a*b[i]
            M[i,:]=M[i,:] - a*M[k,:]
    for i in range (1,n+1):
        x[-i] = (b[-i]-np.dot(M[-i,:],x[:]))
    return x
```

```python
def inversa(M):
    n=len(M[1,:])
    I=np.identity(n)

    #Triangularizar
    for k in range(0,n):
        M[k,:]=M[k,:]/M[k,k]
        for i in range(n-1,0+k,-1):
            a=M[i,k]
            M[i,:]=M[i,:] - a*M[k,:]
            I[i,:]=I[i,:] - a*I[k,:]
    
    #Acabar de tirar os elementos acima da diagonal
    for q in range(1,n):
        for t in range(n-1-q,-1,-1):
            a=M[t,-q]
            M[t,:]=M[t,:] - (a)*M[-q,:]
            I[t,:]=I[t,:] - (a)*I[-q,:]

    return I
```

```python
#Constantes
k1=k3=k4=10
k2=k5=20

w1=w3=2*50
w2=50

#definição da matriz e do array independente
K=np.array([[60,-10,-20],[-k3,k3+k4,-k4],[-k5,-k4,k4+k5]])
W=np.array([w1,w2,w3])
n=len(W)
K1=np.linalg.inv(K)
#definição de array incognito
x=np.empty(3)
```

```python
#Resultado esperado
print(np.dot(K1,W))
print()
#Invoca-se a função de eliminação de Gauss para resolver o sistema enunciado
print(Gauss(K,W))
print()
#Tentativa de calcular a inversa
print(np.dot(inversa(K),W))

#Não tive tempo de fazer debug, mas tentei fazer a inversa e multiplicar pelo array independente e também fazer eliminação de 
#Gauss
```

```python
def Vin(t):
    if int(2*t)%2==0:
        return 1
    else:
        return -1
```

```python
def equacao(t,Vout,RC):
    return (Vin(t)-Vout)/RC
```

```python
def Euler(func,tempo,x0,RC):
    h=(tempo[-1]-tempo[0])/len(tempo)
    x=[]
    x.append(x0)
    xint=x0
    for i in tempo:
        xint+=h*equacao(i,xint+h,RC)
        x.append(xint)
    return x
```

```python
#Definir a equação diferencial
def funcao(r,t):
    a=1
    b=g=0.5
    d=2
    x=r[0]
    y=r[1]
    fx=a*x-b*x*y
    fy=g*x*y-d*y
    return np.array([fx,fy])
```

```python
#definir o metodo de resolução, optei pelo Runge-Kutta de ordem 4
def RK4(func,tempo,x0):
    h=(tempo[-1]-tempo[0])/len(tempo)
    xf=np.array(x0)
    xint=x0
   
    for i in tempo:
        k0=h*func(xint,i)
        k1=h*func(xint+k0/2,i+h/2)
        k2=h*func(xint+k1/2,i+h/2)
        k3=h*func(xint+k2,i+h)
        xint=xint+(k0+2*k1+2*k2+k3)/6
        xf=np.vstack((xf,xint))
    return xf
```

```python
#plot dos gráficos das populações, tendo em conta que as populações são 100 vezes maiores que x e y
tempo=np.linspace(0,20,10000)
condicoes=np.array([2,2])
#predadores
plt.plot(tempo,100*RK4(funcao,tempo,condicoes)[1:,1],'r')
#presas
plt.plot(tempo,100*RK4(funcao,tempo,condicoes)[1:,0],'b')
plt.show()
#Verifica-se um aumento substancial de presas nos tempos em que as populações de predadores entram em declinio
#Quando as populações das presas são máximas, o número de predadores aumenta
#Ao aumentarem os predadores, as presas diminuem
#O ciclo reinicia
```

```python
M=100 #grelha 100x100
target=1e-6 #criterio de convergencia
h=1e-10 #passo da derivada, ou seja, fator que multiplica o rho ao rearranjar os termos da equação de Poisson com a expressão das 
#segundas derivadas

#Ao criar o array a zeros já tem lá as condições de fronteira da caixa
phi=np.zeros([M+1,M+1])

#condições de fronteira para as cargas
phi[int(M/5):int(2*M/5),int(3*M/5):int(4*M/5)]=1
phi[int(3*M/5):int(4*M/5),int(M/5):int(2*M/5)]=-1

#Cria-se um array rho para usar no calculo do potencial
rho=np.zeros([M+1,M+1])
rho[int(M/5):int(2*M/5),int(3*M/5):int(4*M/5)]=1
rho[int(3*M/5):int(4*M/5),int(M/5):int(2*M/5)]=-1

delta=1
w=0.95
while delta>target:
    oldphi=np.copy(phi)
    
    #Forçam
    phi[int(M/5):int(2*M/5),int(3*M/5):int(4*M/5)]=1
    phi[int(3*M/5):int(4*M/5),int(M/5):int(2*M/5)]=-1
    #Vetoriza-se para ser mais eficiente
    phi[1:M,1:M]=w*(phi[2:M+1,1:M]+phi[0:M-1,1:M]+phi[1:M,2:M+1]+phi[1:M,0:M-1]+h*h*rho[1:M,1:M])/4 +(1-w)*phi[1:M,1:M]
    
    delta=np.max(abs(phi-oldphi))
```

```python
plt.imshow(phi)
plt.colorbar()
plt.show()
#Apresenta-se o gráfico de densidade, que se comporta como esperado
```
