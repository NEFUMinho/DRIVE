c) Force os √∫ltimos 90% dos elementos do array a ser zero e calcule
a transformada de Fourier inversa usando a fun√ß√£o irfft. Represente o resultado obtido, sobrepondo-o ao gr√°fico anterior

```
print(len(coeficientes))
coeficientes2=fft.rfft(dados2)

n1=int(len(coeficientes2)*0.5)
coeficientes2[n1:]=0.

inversacoef1=fft.irfft(coeficientes2)

plt.plot(inversacoef1)
plt.plot(coeficientes)
```

```
n1=int(len(dados2)*0.5)
dados2[n1:]=0

inversadados2=fft.irfft(dados2)

plt.plot(inversadados2)
plt.plot(dados)
```

d) Repita o exerc√≠cio anterior, mas desta vez usando apenas os primeiros 2% dos coeficientes da transformada de Fourier.

```
coeficientes3=fft.rfft(dados2)

n2=int(len(coeficientes3)*0.2)
coeficientes3[:n2]=0.

inversacoef3=fft.irfft(coeficientes3)

plt.plot(inversacoef3)
plt.plot(coeficientes)
```
