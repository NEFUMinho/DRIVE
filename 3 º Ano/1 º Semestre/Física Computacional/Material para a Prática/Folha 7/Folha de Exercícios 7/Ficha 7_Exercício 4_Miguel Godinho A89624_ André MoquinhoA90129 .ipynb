```python
#Ficha 7
#Exercício 4
#Exercício para casa
#Miguel Godinho A89624 e André Mouquinho A90129

#import necessários.
from numpy import arange,array
import pylab #podia só dar import ao necessário, mas como já tinha feito o programa assim seria uma perda de tempo alterar.


#primeiro vou definir um vetor v com as nossas três váriaveis dependentes, x,y,z logo v=[x,y,z].
#então fiz uma função do nosso vetor depende do tempo:
def f(v,t):
    #primeiro as constante, sigma, r e b.
    s=10 #sigma
    r=28 
    b=8/3
    #agora definimos as posições das nossas variáveis dependentes no array (vetor) v.
    x= v[0]
    y= v[1]
    z= v[2]
    #e vamos buscar os valores das nossas funções fx,fy,fz
    fx= s*(y - x)
    fy= r*x - y - x*z
    fz= x*y - b*z
    #terminamos por devolver um array float com os valores de fx, fy e fz
    return array([fx,fy,fz],float)


    #vou fazer agora uma função para determinar os valores de t,x,y,z dependente do valor N, aqui podemos dizer que t tbm é dependente, mas analitcamente não o é.
def d(N):
    #os t incial e final são dados pelo enunciado
    a=0.0 #tinicial
    b=50.0 #tfinal
    #valor de h
    h=(b-a)/N
    #condições iniciais para t=0
    r=array([0.0,1.0,0.0],float)
    #fiz outro array float
    tpoints=arange(a,b,h)
    #os valores de t, vão ser de a a b com um "passo" de h
    xpoints= []
    ypoints= []
    zpoints= []
    #arrays para x,y,z
    #agora vou fazer um ciclo para determinar os valores de x,y,z para diferentes t
    for t in tpoints:
        xpoints.append(r[0])
        ypoints.append(r[1])
        zpoints.append(r[2])
        #começo por adicionar os valores inicais para t=0
        #vou fazer em quarta ordem, porque já foi provado na aula que é mais preciso.
        k1 = h*f(r,t)
        k2 = h*f(r+0.5*k1,t+0.5*h)    
        k3 = h*f(r+0.5*k2,t+0.5*h)
        k4 = h*f(r+k3,t+h)
        r += (k1+2*k2+2*k3+k4)*(1/6)
        #não é nada de novo, é o método de Runge-Kuta para a quarta ordem.
        #aqui quando o ciclo reencia temos valores diferentes de r e vão ser adicionados antes.
        # e dou return nos arrays de t,x,y,z
    return tpoints,xpoints,ypoints,zpoints

    #gora vou correr os gráficos

for i in range (1,3):
    #vou dar plot para diferentes N e vou subir N em exponencial de 10 de 1 até 3.
    #vou fazer todos os gráficos de para cada N
    N=10**i
    #x em funão de t
    pylab.plot(d(N)[0],d(N)[1], label=N)
    pylab.xlabel("t")
    pylab.ylabel("x(t)")
    pylab.title("Gráficos x em função de t")
    pylab.legend()
    pylab.show()
    #y em funão de t
    pylab.plot(d(N)[0],d(N)[2], label=N)
    pylab.xlabel("t")
    pylab.ylabel("y(t)")
    pylab.title("Gráficos y em função de t")
    pylab.legend()
    pylab.show()
    #z em funão de t
    pylab.plot(d(N)[0],d(N)[3], label=N)
    pylab.xlabel("t")
    pylab.ylabel("z(t)")
    pylab.title("Gráficos z em função de t")
    pylab.legend()
    pylab.show()
    #z em funão de x
    pylab.plot(d(N)[1],d(N)[3], label=N)
    pylab.xlabel("x")
    pylab.ylabel("z(x)")
    pylab.title("Gráficos z em função de x")
    pylab.legend()
    pylab.show()
    #aqui conseguimos verificar que o N é demasiado pequeno por motivos matemáticos que não importam e não nos serve.
    #Até indica mesmo quando rodamos o programa que temos um erro.

    
for i in range (4,7):
    #vou dar plot para diferentes N e vou subir N em exponencial de 10 de 4 a 6.
    #vou fazer todos os gráficos de para cada N
    N=10**i
    #x em funão de t
    pylab.plot(d(N)[0],d(N)[1], label=N)
    pylab.xlabel("t")
    pylab.ylabel("x(t)")
    pylab.title("Gráficos x em função de t")
    pylab.legend()
    pylab.show()
    #y em funão de t
    pylab.plot(d(N)[0],d(N)[2], label=N)
    pylab.xlabel("t")
    pylab.ylabel("y(t)")
    pylab.title("Gráficos y em função de t")
    pylab.legend()
    pylab.show()
    #z em funão de t
    pylab.plot(d(N)[0],d(N)[3], label=N)
    pylab.xlabel("t")
    pylab.ylabel("z(t)")
    pylab.title("Gráficos z em função de t")
    pylab.legend()
    pylab.show()
    #z em funão de x
    pylab.plot(d(N)[1],d(N)[3], label=N)
    pylab.xlabel("x")
    pylab.ylabel("z(x)")
    pylab.title("Gráficos z em função de x")
    pylab.legend()
    pylab.show()
    #aqui já obtemos o esperado. Ou seja pelo menos com N=10**4 já conseguimos obter o que queremos
    
    
#Observação Final: fizemos uma investigação e o erro deve-se a constante de Lipschitz, L. 
#Para este método, os passos utéis dependem da constante de Lipshitz, em que L*h deve ter valores no intervalo de 1e-3 até 0.1
#E sabemos tbm que para as equações de Lorenz L=50.
```
