```python
#Exercício 7
import numpy as np
import pylab
def fact(y):
    if np.any(y == 0):
        fact=1
        return fact
    else:
        fact=1
        for k in range (1, y+1):
            fact = fact * k
        return fact

def e_negativo_x(k):
    n=0
    precisao_numerica=1
    precisao_numerica_ideal= 10**(-8)
    if np.any(k<20):
        while np.any(precisao_numerica > precisao_numerica_ideal):
            som=0
            for i in range (0, n+1):
                ind=((-k)**i)/fact(i)
                if np.any(i> 0 & n-i == 0):
                    ind_2=((-k)**(i-1))/fact(i-1)
                    precisao_numerica_1 = ind_2 - ind
                    precisao_numerica = abs(precisao_numerica_1)
                som = som + ind
            if   np.any(precisao_numerica > precisao_numerica_ideal):
                n += 1
    elif np.any(k>=20):
        k=k/10
        while np.any(precisao_numerica > precisao_numerica_ideal):
            som=0
            for i in range (0, n+1):
                ind=((-k)**i)/fact(i)
                if np.any(i> 0 & n-i == 0):
                    ind_2=((-k)**(i-1))/fact(i-1)
                    precisao_numerica_1 = ind_2 - ind
                    precisao_numerica = abs(precisao_numerica_1)
                som = som + ind
            if   np.any(precisao_numerica > precisao_numerica_ideal):
                n += 1
        som=som**10
    return som



print(e_negativo_x(0.1))
print(e_negativo_x(1))
print(e_negativo_x(10))
print(e_negativo_x(15))
print(e_negativo_x(30))
                                                                 
print(np.exp(-0.1))
print(np.exp(-1))
print(np.exp(-10))
print(np.exp(-15))
print(np.exp(-30))


x = np.linspace(0,10,100)
y = e_negativo_x(x)
pylab.plot(x,y)
pylab.show()
y = np.exp(-x)
pylab.plot(x,y)
pylab.show()
```

```python
#exercício 8
from math import pi, sqrt
from numpy import linspace, exp, vectorize
import pylab
def f(x):
    t=exp(-pi*x)
    return (1-sqrt(1-t**2))/(t)

def f_2(x):
    t=exp(-pi*x)
    return (t)/(1+sqrt(1-t**2))

print(f(6))
print(f_2(6))
x=linspace(5,7,100)
y_2=vectorize(f)
y=y_2(x)
pylab.plot(x,y)
pylab.show()
y_2=vectorize(f_2)
y=y_2(x)
pylab.plot(x,y)
pylab.show()
```

```python
#exercício 9
import pylab
from numpy import linspace, vectorize
#a)
def s(N):
    som_up=0
    som_down=0
    for i in range (1,N+1):
        som_up += 1/(i)
    for i in range (N,0,-1):
        som_down += 1/(i)
    return som_up, som_down
        
N=int(input("Escolha o número de N:"))

print(s(N)[0])
print(s(N)[1])

#b)
x=[]
for i in range (1,1000):
    x.append(i)
s_2=vectorize(s)
y=(s_2(x)[0]-s_2(x)[1])/(s_2(x)[0]+s_2(x)[1])
pylab.plot(x,y)
pylab.show()
```
