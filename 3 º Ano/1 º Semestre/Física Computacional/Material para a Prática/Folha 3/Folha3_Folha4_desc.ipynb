### Exercício 3 - Bisection and Secant 

```python
#Exercício 3 - Bisection
import numpy as np
import matplotlib.pyplot as plt

def bisection(f,a,b,N):
    '''
    Inputs
    f: e a funçao f(x) que estamos a tentar aproximar as soluçoes f(x)=0
    a,b : intervalos onde queremos encontrar a soluçao
    N: numero de interaçoes a implementar
    '''
    if f(a)*f(b) >= 0:
        print("Método da bisseção falha")
        return None
    a_N = a
    b_N = b
    for n in range(1,N+1):
        m_N = (a_N + b_N)/2
        f_m_N = f(m_N)
        if f(a_N)*f_m_N < 0:
            a_N = a_N
            b_N = m_N
        elif f(b_N)*f_m_N < 0:
            a_N = m_N
            b_N = b_N
        elif f_m_N == 0:
            print("Solução exata encontrada:")
            return m_N
        else:
            print("Método da bisseção falha")
            return None
    return (a_N + b_N)/2

x = np.linspace(-2,2,10000)

f = lambda x: (4**x)-(3**(2*x))+(2**(3*x))-1

plt.figure()
plt.plot(x,f(x))
plt.show()

bisection(f,1.5,2,100)
```

```python
#Exercício 3 - Secant
import numpy as np
import sys
import matplotlib.pyplot as plt

def secant(f,a,b,N):
    '''
    Inputs
    f: e a funçao f(x) que estamos a tentar aproximar as soluçoes
    a,b : intervalos onde queremos encontrar a soluçao
    N: numero de interaçoes a implementar
    '''
    if f(a)*f(b) >= 0:
        print("Método da secante falha")
        return None
    a_N = a
    b_N = b
    for n in range(1,N+1):
        m_N = a_N - f(a_N)*(b_N - a_N)/(f(b_N) - f(a_N))
        f_m_N = f(m_N)
        if f(a_N)*f_m_N < 0:
            a_N = a_N
            b_N = m_N
        elif f(b_N)*f_m_N < 0:
            a_N = m_N
            b_N = b_N
        elif f_m_N == 0:
            print("Solução exata encontrada:")
            return m_N
        else:
            print("Método da secante falha")
            return None
    return a_N - f(a_N)*(b_N - a_N)/(f(b_N) - f(a_N))

x = np.linspace(-2,2,100)

f = lambda x:(4**x)-(3**(2*x))+(2**(3*x))-1

plt.figure()
plt.plot(x,f(x))
plt.show()

secant(f,1.5,2,1000)
```

### Exercicio 4 - Derivada, Metodo Newton 

```python
import matplotlib.pyplot as plt
import numpy as np


def newton(f,df,x0,epsilon,max_iter):
    xn = x0
    for n in range(0,max_iter):
        fxn = f(xn)
        if abs(fxn) < epsilon:
            print('Soluçao encontrada depois de ',n,'interaçoes.')
            return xn
        dfxn = df(xn)
        if dfxn == 0:
            print('0 derivadas. Nao foi encontrada a soluçao.')
            return None
        xn = xn - fxn/dfxn
    print('Passou o numero maximo de interaçoes. Nao foi encontrada soluçao.')
    return None

x = np.linspace(-10,10,100)

f = lambda x: x**4-x-1
df = lambda x: 4*x**3-1
ddf = lambda x: 12*x**2
dddf = lambda x: 24*x

plt.figure()
plt.plot(x,f(x),label ="f")
plt.plot(x,df(x),label ="df")
plt.plot(x,ddf(x),label ="ddf")
plt.plot(x,dddf(x),label ="dddf")
plt.legend()
plt.show()

newton(f,df,1,1e-8,100)
```

# Exercicio 5 - Gaussian Eliminação (sistema linear)

```python
#TPC
from numpy import array,empty

A = array([[ 3,  2],
           [ 1,  1],], float)
v = array([ 5, 3],float)
N = len(v)

# Gaussian elimination
for m in range(N):

    # Divide by the diagonal element
    div = A[m,m]
    A[m,:] /= div
    v[m] /= div

    # Now subtract from the lower rows
    for i in range(m+1,N):
        mult = A[i,m]
        A[i,:] -= mult*A[m,:]
        v[i] -= mult*v[m]

# Backsubstitution
x = empty(N,float)
for m in range(N-1,-1,-1):
    x[m] = v[m]
    for i in range(m+1,N):
        x[m] -= A[m,i]*x[i]

print(x)
```
