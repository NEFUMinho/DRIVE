# Movimento obliquo com atrito proporcional √° densidade do ar e quadrado da velocidade

```python
#Fun√ß√£o da densidade do ar
def T(y):
    To = 288.15
    L = 0.0065
    return To-(L*y)
def p(y):
    po = 101325
    L = 0.0065
    To = 288.15
    g = 9.8
    M = 0.0289644
    R = 8.31447 
    return po*((1-((L*y)/To))**((g*M)/(R*L)))
def rho_(y):
    M = 0.0289644
    R = 8.31447 
    return ((p(y)*M)/(R*T(y)))

def rho(y):
    po = 101325
    L = 0.0065
    To = 288.15
    g = 9.8
    M = 0.0289644
    R = 8.31447 
    rhoo = po*M/R*To
    Hn = 1/(((g*M)/(R*To))-(L/To))
    return rhoo*exp(-y/Hn)
```
