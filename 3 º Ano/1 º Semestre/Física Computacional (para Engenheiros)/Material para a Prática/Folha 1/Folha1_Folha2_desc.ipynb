```python
#Exercico 2
#Exercicio n.2 a)
import matplotlib.pyplot as plt

def Sup(N):
    Sup=0
#    Aup = []
    if N<=0:
        print ('impossible')
    else:
        for n in range (1,int(N+1)):
            Sup=Sup+(1/n)
#            Aup.append(n)
#            print(n)
    return (Sup)

print(Sup(5))
print('___________')

def Sdown(N):
    Sdown=0
#    Adown = []
    if N<=0:
        print ('impossible')
    else:
        for i in range (0, int(N)):
            k=N-i
            Sdown=Sdown+float(1/k)
#            Adown.append(Sdown)
#            print(k)
    return (Sdown)

print(Sdown(5))
print('___________')

X=np.linspace(1,10)
f_up=[Sup(x) for x in X]
f_down=[Sdown(x) for x in X]

plt.plot(x, f_up)
plt.plot(x, f_down)
plt.xlabel('x')
plt.ylabel('y')

plt.loglog(1,Sup(5))
plt.loglog(1,Sdown(5))
plt.show()
```
