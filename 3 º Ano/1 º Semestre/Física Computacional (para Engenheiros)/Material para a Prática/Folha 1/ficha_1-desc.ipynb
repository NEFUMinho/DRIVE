Exercicio 1 - Nº primos menores que 10000

```python
x=[]
for i in range(1,10000):
    div=0
    for a in range(2,i):
        if i % a == 0:
            div=div+1
        if div!=0:
            break
    if div == 0:
        x.append(i)
print(x)
```

Exercicio 5 - Representação Gráfica 3D

```python
def f(x,y):
    return np.sin(x)*np.cos(y)

fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-3,3)
y = np.linspace(-3,3)
X,Y = np.meshgrid(x,y)
Z = f(X,Y)
print(X,Y)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap='viridis',edgecolor='none')
```

Exercicio 7 - Função e^-x

```python
def fact(x): #função fatorial
    if x==0:
        return 1
    else: 
        f=1
        for i in range(1,x+1):
            f=i*f
        return f
    
def mod(x):
    if x>=0:
        return x
    if x<0:
        return -1*x
    
def aexp(x): #função exponencial de -x
    e=0
    for i in range(10000000):
        b=e
        a=pow(-1,i)*pow(x,i)/fact(i)
        if b==0:
            e=a+b
        elif mod(a)/mod(b)<=pow(10,-8):
            break
        else: 
             e=a+b
    return e
x=int(input('Escreva o seu x: '))
r=np.exp(-x)
a=aexp(x)
e=a/r
print('Valor aproximado: ',a)
print('Valor real: ',r)
print('O erro é de: ',e)
```

```python
x = np.linspace(-3,10)

plt.subplot(221)
plt.plot(x,np.exp(-x), 'g')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Real')

plt.subplot(222)
f_list=[aexp(a) for a in x] #passa de array para valores individuais, para ser usado na função aexp
plt.plot(x,f_list, 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Aproximation')
plt.tight_layout()
```

Exercício 8 - Calculo do Pi

```python
N=int(input('Número de pontos a gerar: '))
Nd=0 #pontos dentro do circulo
Nf=0 #pontos fora do circulo
for i in range (1,N+1):
    x=rand.random()
    y=rand.random()
    z=np.sqrt((0.5-x)**2+(0.5-y)**2)
    if z<=0.5:
        Nd+=1
    else: 
        Nf+=1
        
Pi=4*Nd/(Nd+Nf)
print('O valor aproximado de Pi é ',Pi)
print('O valor real de Pi é ',np.pi)

    
```
