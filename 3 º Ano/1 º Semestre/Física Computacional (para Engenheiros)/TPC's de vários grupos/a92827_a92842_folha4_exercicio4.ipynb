(a) Mostre que o sistema é descrito pelo seguinte sistema de equações, onde α = 2k − mω2:
$ \\
(α − k)x_1 − kx_2 = C \\
αx_i − kx_i-1 − kx_i+1 = 0 \\
(α − k)x_N − kx_N−1 = 0 \\
$

Sendo:  
$ \\
ξ_{i}(t)' = iwx_{i}e^{iwt} $

$ξ_{i}(t)'' = -w^{2}x_{i}e^{iwt} $

Substituímos os valores das derivadas nas equações:
$ \\
m{d^{2}ξ_{i} \over dt^{2}} = k(ξ_{i+1} - ξ_{i}) + k(ξ_{i-1} - ξ_{i}) + F_{i} (1)\\
m{d^{2}ξ_{i} \over dt^{2}} = k(ξ_{2} - ξ_{1}) + F_{i} (2)\\
m{d^{2}ξ_{N} \over dt^{2}} = k(ξ_{N-1} - ξ_{N}) + F_{N} (3) $

Substituindo os valores em (1) e sabendo que nenhuma força está a ser aplicada às massas intermédias, temos:

$ \\
-mw^{2}x_{i}e^{iwt} = kx_{i+1}e^{iwt}-kx_{i}e^{iwt} + kx_{i-1}e^{iwt} -kx_{i}e^{iwt} <=> $

$ <=> -mw^{2}x_{i} = kx_{i+1} -2kx_{i} + kx_{i-1} <=> $

$<=> (2k-mw^{2})x_{i} -kx_{i+1} - kx_{i-1} = 0 $

$ <=> αx_{i} - kx_{i+1} - kx_{i-1} = 0 $

Substituindo os valores em (2) e sabendo que a força aplicada à massa 1 é do tipo $ F_1=ce^{iwt}$, temos:

$ \\
-mw^{2}x_{1}e^{iwt} = k(x_{2}e^{iwt}-x_{1}e^{iwt}) + ce^{iwt} <=> $

$ <=> -mw^{2}x_{1} = k(x_{2}-x_{1}) + c <=> $

$ <=> -mw^{2}x_{1} - kx_{2} +kx_{1} = c <=> $

$<=> (k - mw^{2})x_{1} -kx_{2} = c $

Substituindo os valores em (3) e sabendo que nenhuma força está a ser aplicada à massa N, temos:

$ \\
-mw^{2}x_{N}e^{iwt} = k(x_{N-1}e^{iwt}-x_{N}e^{iwt}) <=> $

$ <=> -mw^{2}x_{N} = k(x_{N-1}-x_{N}) <=> $

$ <=> -mw^{2}x_{N} - kx_{N-1} +kx_{N} = 0 <=> $

$<=> (k - mw^{2})x_{N} -kx_{N-1} = 0 $


(b) Resolva este problema para N = 26, C = 1, m = 1, k = 6 e ω = 2.

```python
import numpy as np
def eq_to_mat(N,C,m,k,w):                                   #definimos a equação que recebe os nossos dados e cria uma matriz
    a = 2*k-m*(w**2)                                        #definimos a variável alpha
    lista = []                                              #definimos 'lista', que será a nossa matriz final
    listt = []                                              #definimos 'listt', que ajudará a criar cada linha da matriz    
    lista.append([a-k,-k])                                  #adicionamos a lista [a-k,-k] a 'lista', que são os primeiros coeficientes da primeira equação
    for b in range(N-2):                                    #os restantes coeficientes (ou seja, N-2) são zeros
        lista[0].append(0)                                  #adicionamos ao primeiro elemento de 'lista' (lista[0]) os zeros em falta
    lista[0].append(C)                                      #por fim, adicionamos C, a constante à qual a equação está igualada
    for i in range(1,N-1):                                  #na próxima porção de código, adicionamos a 'lista' as linhas compreendidas entre a primeira e a última linhas
        for b in range(0,i-1):                              #adicionamos os 0 que estão antes dos coeficientes em questão (depende da linha a ser trabalhada)
            listt.append(0)
        listt.append(-k)                                    #adicionamos o coeficiente de x_i-1
        listt.append(a)                                     #adicionamos o coeficiente de x_i
        listt.append(-k)                                    #adicionamos o coeficiente de x_i+1
        for b in range(i+1,N):                              #adicionamos os 0 que estão depois dos coeficientes em questão (de i+1 até N, porque a equação está igualada a 0, então adicionamos a última constante)
            listt.append(0)
        lista.append(listt)                                 #depois de criada a linha da matriz, esta é adicionada à lista que representa a matriz
        listt = []                                          #voltamos a igualar listt a [], para que, ao percorrer novamente o ciclo, possamos criar a linha seguinte
    for b in range(N-2):                                    #na próxima porção de código vamos fazer a última linha da matriz
        listt.append(0)                                     #adicionamos os zeros antes dos coeficientes em questão
    listt.append(-k)                                        #adicionamos o coeficiente de x_N-1
    listt.append(a-k)                                       #adicionamos o coeficiente de x_N
    listt.append(0)                                         #adiconamos o 0 à qual a equação está igualada
    lista.append(listt)                                     #adicionamos a 'lista' a última coluna da matriz
    return lista                                            #a função devolve a matriz

#Teste da função
print(np.array(eq_to_mat(26,1,1,6,2)))
```

```python
def metodo(A):                                                 #definimos a função que vai aplicar o método de gauss
    solucoes=[]                                                #criamos uma lista na qual vão ser armazenadas as soluções
    for i in range(len(A)):                                    #a variável i vai dar a linha em que estamos a trabalhar
        for m in range(i+1, len(A)):                           #a variável m vai dar a linha abaixo de i       
            mult = A[m][i]/A[i][i]                             #calculamos o fator multiplicativo do método de gauss, dado por a_mn/a_nn, sendo a_nn o elemento pivô de cada linha e a_mn o elemento imediatamente abaixo dele                                        
            for n in range (i, len(A)+1):                      #a variável n vai percorrer todos os elementos de cada linha a partir de i
                A[m][n] = A[m][n]- mult*A[i][n]                #aplicamos a transformação L_m = L_m - (a_mn/a_nn)*L_n a cada elemento da respetiva linha (seguindo a terminologia dos comentários e não do código), por forma a obter uma matriz triangular
    
    solucoes = np.zeros(len(A))                                #vai armazenar na lista vazia o número de soluções do sistema (igual ao número de linhas da matriz) enquanto zeros
    linha = len(A)-1                                           #a variável linha guarda a linha em que estamos a trabalhar
    b = len(A)                                                 #a variável b guarda o número da última coluna, que contém os números aos quais as expressões estão igualadas 
    while linha>=0:                                            #como vamos decrementar 'linha', impomos esta condição. começamos o ciclo na última linha
        x = A[linha][b]                                        #atribuimos à variável x o valor contido na última coluna, na respetiva linha a ser trabalhada
        coluna = len(A)-1                                      #a variável coluna guarda a coluna em que estamos a trabalhar (excluímos a última porque não é trabalhada)
        while coluna>linha:                                    #entramos no ciclo se estivermos a trabalhar numa posição da matriz acima da diagonal criada na triangularização. no primeiro caso a ser trabalhado (última linha e penúltima coluna), não entramos no ciclo porque linha>coluna, armazenando logo o valor de x_m na última posição da lista solucoes
            x = x - A[linha][coluna]*solucoes[coluna]          #recorrendo à fórmula do método de Gauss
            coluna = coluna -1                                 #decrementamos o valor da coluna em 1
        x = x/A[linha][linha]                                  
        linha = linha -1                                       #decrementamos o valor da linha em 1 para trabalhar a equação da linha anterior
        solucoes[coluna]=x                                     #igualamos o valor da solução para a coluna em que estamos a trabalhar a x.

    return solucoes
```

```python
from numpy import copy
#Alteramos a primeira função, de forma a que a última coluna não exista, para que sirva de argumento à função banded

def eq_to_mat2(N,C,m,k,w):                                  #definimos a equação que recebe os nossos dados e cria uma matriz
    a = 2*k-m*(w**2)                                        #definimos a variável alpha
    lista = []                                              #definimos 'lista', que será a nossa matriz final
    listt = []                                              #definimos 'listt', que ajudará a criar cada linha da matriz    
    lista.append([a-k,-k])                                  #adicionamos a lista [a-k,-k] a 'lista', que são os primeiros coeficientes da primeira equação
    for b in range(N-2):                                    #os restantes coeficientes (ou seja, N-2) são zeros
        lista[0].append(0)                                  #adicionamos ao primeiro elemento de 'lista' (lista[0]) os zeros em falta

    for i in range(1,N-1):                                  #na próxima porção de código, adicionamos a 'lista' as linhas compreendidas entre a primeira e a última linhas
        for b in range(0,i-1):                              #adicionamos os 0 que estão antes dos coeficientes em questão (depende da linha a ser trabalhada)
            listt.append(0)
        listt.append(-k)                                    #adicionamos o coeficiente de x_i-1
        listt.append(a)                                     #adicionamos o coeficiente de x_i
        listt.append(-k)                                    #adicionamos o coeficiente de x_i+1
        for b in range(i+1,N-1):                            #adicionamos os 0 que estão depois dos coeficientes em questão (de i+1 até N-1, porque não queremos a última coluna)
            listt.append(0)
        lista.append(listt)                                 #depois de criada a linha da matriz, esta é adicionada à lista que representa a matriz
        listt = []                                          #voltamos a igualar listt a [], para que, ao percorrer novamente o ciclo, possamos criar a linha seguinte
    for b in range(N-2):                                    #na próxima porção de código vamos fazer a última linha da matriz
        listt.append(0)                                     #adicionamos os zeros antes dos coeficientes em questão
    listt.append(-k)                                        #adicionamos o coeficiente de x_N-1
    listt.append(a-k)                                       #adicionamos o coeficiente de x_N

    lista.append(listt)                                     #adicionamos a 'lista' a última linha da matriz
    
    return lista                                            #a função devolve a matriz
 

def banded(Aa,va,up,down):

    # Copy the inputs and determine the size of the system
    A = copy(Aa)
    v = copy(va)
    N = len(v)

    # Gaussian elimination
    for m in range(N):

        # Normalization factor
        div = A[up,m]

        # Update the vector first
        v[m] /= div
        for k in range(1,down+1):
            if m+k<N:
                v[m+k] -= A[up+k,m]*v[m]

        # Now normalize the pivot row of A and subtract from lower ones
        for i in range(up):
            j = m + up - i
            if j<N:
                A[i,j] /= div
                for k in range(1,down+1):
                    A[i+k,j] -= A[up+k,m]*A[i,j]

    # Backsubstitution
    for m in range(N-2,-1,-1):
        for i in range(up):
            j = m + up - i
            if j<N:
                v[m] -= A[i,j]*v[j]

    return v

#A função fun_f será a função final à qual será dada apenas a função obtida em eq_to_mat2. Esta terá como tarefa criar e atribuír as variáveis à banded.

def fun_f(A):
    #denominamos as variáveis respectivas a este problema.
    k = 6
    m = 1
    w = 2
    a = 2*k-m*(w**2)
    N = len(A)                                              #N é igual ao tamanho de A
    
#Críamos a lista que será dada na função banded como o argumento 'va'. Esta lista é composta N elementos em que o primeiro é igual a C e os restantes a zero (neste problema, C=1) e é equivalente
#às igualdades das equações iniciais.

    v = np.zeros(N,float)                                   #críamos lista com N elementos iguais a 0.
    v[0] = 1                                                #igualamos o primeiro elemento da lista a C que para este problema é 1.

#a função banded recebe uma matriz composta pelas 3 diagonais com elementos diferentes de 0 que compóem a matriz obtida na função eq_to_mat2 excluindo os 0 desnecessários aos cálculos.
#para este caso:

    A = np.empty([3,N],float)                               #cria-se 3 listas com N elementos  
    A[0,:] = -k                                             #os primeiros elementos de cada lista em A é igualado a -k
    A[1,:] = a                                              #os segundos elementos de cada lista em A é igualado a a
    A[2,:] = -k                                             #os terceiros elementos de cada lista em A é igualado a -k
    A[1,0] = a -k                                           #os elementos das extremidades das matrizes sabemos pelas equações que o primeiro elemento da segunda linha é igual a a-k
    A[1,N-1] = a -k                                         #o último elemento da segunda linha é igual a a-k

    resultado = banded(A,v,1,1)                             #o resultado final será dado pela função banded.
    
    return resultado

#Teste da função fun_f:
fun_f(eq_to_mat2(1000,1,1,6,2))

#os resultados são apresentados imediatamente a seguir ao pedido de execução da célula provando a melhor eficácia comparativamente ao método anterior.
```
