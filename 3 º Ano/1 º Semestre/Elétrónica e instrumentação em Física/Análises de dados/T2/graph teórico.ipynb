```python
from math import sqrt,log10
from pylab import plot,show,title,xlabel,ylabel,savefig
from numpy import linspace,zeros
x=linspace(0,10,10)
y=[]
for i in range (0,10):
    z=20*log10(1/(sqrt(1 + (x[i]/3.302)**2)))
    y.append(z)
plot(x,y)
title('Função de Transferência')
xlabel('Frequência (kHz)')
ylabel('Magnitude do Ganho (dB)')
savefig('g3.png')
show()



 
```

```python
from math import sqrt,log10, atan
from pylab import plot,show,title,xlabel,ylabel,savefig, degrees
from numpy import linspace,zeros
x=linspace(0,10,10)
y=[]
for i in range (0,10):
    f=atan(-x[i]/3.302)
    z=degrees(f)
    y.append(z)
plot(x,y)
title('Função de Transferência')
xlabel('Frequência (kHz)')
ylabel('Fase (Degree)')
savefig('g4.png')
show()
```
